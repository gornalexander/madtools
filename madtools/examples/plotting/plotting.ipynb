```python
import madtools.madplotter as mp
import madtools.madutils as mu
import madtools.madtracker as mt

import pandas as pd
from cpymad.madx import Madx
import requests
import holoviews as hv

hv.extension('bokeh')
```

```python
seq = requests.get('https://gitlab.cern.ch/acc-models/acc-models-tls/-/raw/post_ls3/sps_extraction/tt20bdf_unsplit/tt20bdf.seq').text
inp = requests.get('https://gitlab.cern.ch/acc-models/acc-models-tls/-/raw/main/sps_extraction/tt20t2/tt20_initial_conditions.inp').text
seq = "\n".join(line for line in seq.splitlines() if "end_tt24" not in line) # remove unsupported code

def init_madx():
    madx = Madx(stdout=False)
    madx.input(inp)
    madx.input(seq)

    madx.select(flag='twiss', column=['s', 'l', 'betx', 'bety', 'betx', 'bety','dx', 'dy', 'mux','muy', 'N1', 'apertype', 'aper_1', 'aper_2'])
    madx.use(sequence='tt20bdf')
    return madx

madx = init_madx()
twiss = madx.twiss(beta0='INITBETA0_TT20_instant_20mm_4um').dframe() # init conditions from acc-models repo
```

# Plot lattice and twiss functions

```python
lattice = mp.plot_lattice(twiss, show_markers=True, ylim=(-0.1, 0.1), twiss_plots=['beta', 'd'], range_='ejpt/t2center')
lattice
```

# Track the beam and plot envelope

```python
madx = init_madx()

beam_split2 = pd.read_pickle('beam_split2.pkl')
beam_split2['pt'] = beam_split2['delta'] # rename delta to pt for consistency with madx
beam_split2['number'] = range(len(beam_split2.index)) # add a number column for tracking purposes
beam_split2['t'] = 0

init_cond = mu.get_initial_condition(beam_split2) # for full beam parameters list use 'mu.calc_beam_pars(beam_split2)' instead

beam_tracked, tw = mt.transfer_beam(madx, beam_split2, init_cond, sequence='tt24p42', range_='endsplit2/#e')

envelope = mu.calc_envelope(beam_tracked)
```

```python
lattice_and_env = mp.plot_all(tw, envelope, show_markers=True, ylim=(-0.1, 0.1), twiss_plots=['beta', 'd'])
lattice_and_env
```

# Saving plots

```python
hv.save(lattice_and_env, 'lattice_and_env.html') # Might require installing some libraries for pdf and other formats.
```
